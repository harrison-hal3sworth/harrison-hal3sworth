(Big hint: Your answer should incorporate some use of 1.5 * IQR, and note that notebook markdown supports embedded $LaTeX$, such as using $Q1 - 1.5 * IQR$)

**ANSWER:** An outlier in terms of $IQR$ is defined as any value that is $1.5*IQR+Q_3$ or  $Q_1-1.5*IQR$

---
**16)** [P] Write a function called `IQR_outlier_limits` that takes a dataframe as input, and computes the minimum and maximum outlier thresholds for each variable (numeric only assumed), stored in a data frame that is returned as a result.

````python
# ANSWER:

def IQR_outlier_limits(df):
    """ Compute the outlier limits based on the 1.5 * IQR rule
        Input: A DataFrame
        Output: a DataFrame of the minimum and maximum outlier thresholds for all 
                numeric variables
    """
    IQR = df.quantile(0.75, axis=0, numeric_only=True) - df.quantile(0.25, axis=0, numeric_only=True)
    results = pd.DataFrame([df.quantile(0.25, axis=0, numeric_only=True)-(IQR*1.5), (IQR*1.5) + df.quantile(0.75, axis=0, numeric_only=True)], index=["min_out","max_out"])
    results=results[["age","numvisit"]]
    
    return results
````

# Part III - Discretization
---

As we learn about doing EDA and prepping our data, we're learning that visualizing our data to get a good understanding of our variables is important. We're also learning that we can make a variety of transformations with our variables in ways that let us do more meaningful visualizations (and also, as we'll soon understand, modeling).

One common transformation is called **discretization**. It's the process of converting a numeric variable into a categorical variable, more specifically, usually an ordinal.

A common way of referring to categorical variables are **factors**. Factors are explanatory categorical variables in a controlled experiment. They typically only take on a small number of values, called **levels**.

---
**38)**	[P] Create an additional variable in df_badhealth called `age_fac1` that represents a factor with 3 levels, "young", "mid", and "older". You should discretize according to equal width bins over the distribution of the age variable.



````python
# ANSWER
# https://pandas.pydata.org/docs/reference/api/pandas.cut.html

bins = pd.cut(health_df["age"], bins=3, labels=["young", "mid", "older"], include_lowest=True)
health_df["age_fac1"] = bins
health_df.head()
````

---
**39)**	[P] Create an additional variable in df_badhealth called age_fac2 that represents a factor with 3 levels, "young", "mid", and "older". However, this time you should discretize using equal depth bins over the distribution of age. (i.e. the division is chosen to try and get close to an equal number of observations in each bin as possible.) Show the value counts for df_badhealth.age_fac2.





````python
# ANSWER

bins = pd.qcut(health_df["age"], q=3, labels=["young", "mid", "older"])
health_df["age_fac2"] = bins
health_df.head()
````
