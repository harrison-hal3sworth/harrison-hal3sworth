```python
def create_keras_model():
    inputs = Input(shape=(4,))
    
    hidden_layer1 = Dense(16, activation='relu')(inputs)  # Increase number of units and change activation function
    hidden_layer2 = Dense(8, activation='tanh')(hidden_layer1)  # Add an additional hidden layer with different activation function
    
    outputs = Dense(3, activation='softmax')(hidden_layer2)
    
    model = Model(inputs=inputs, outputs=outputs, name="simple_iris_model")
    
    return model
```

24) Now, harness the power of wrapping this class. Use your code from lab10 that performed a full cross
validation. For sake of your time, you may set K to 5. Also, you will likely want to disable verbose mode for
this, otherwise you'll have a LOT of output. AND, because deep learning models can take a while to train each
model, it is a good idea to generate some output in your loop to show that the cross validation is progressing.
For example:

Starting Fold #1

Fold #1 complete, time = 10.15 sec

Starting Fold #2

Fold #2 complete, time = 9.44 sec

Starting Fold #3

Fold #3 complete, time = 10.52 sec

Fold #4 complete, time = 10.04 sec

Starting Fold #5

Fold #5 complete, time = 10.00 sec

```python
k = 5
cv_scores = []
for fold in range(1, k+1):
    print(f"Starting Fold #{fold}")
    fold_start_time = time.time()

    X_train_fold, X_test_fold, y_train_fold, y_test_fold = train_test_split(X_train, y_train_encoded, test_size=0.2)
    clf = KerasClassifier(
        model=create_keras_model,
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy'],
        epochs=100,
        batch_size=1,
        verbose=0
    )
    
    clf.fit(X_train_fold, y_train_fold)
    score = clf.score(X_test_fold, y_test_fold)
    cv_scores.append(score)
    fold_end_time = time.time()
    fold_time = fold_end_time - fold_start_time
    print(f"Fold #{fold} complete, time = {fold_time:.2f} sec")

print("Cross-validation scores:", cv_scores)
print("Mean accuracy:", np.mean(cv_scores))
```
